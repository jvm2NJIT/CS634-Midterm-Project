```python
import random
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

random.seed(2000)
```

This is the list of 10 items that all transactions consist of.

```python
items = sorted(['Shampoo', 'Apple', 'Banana', 'Milk', 'Eggs', 'Soap', 'Bacon', 'Sugar', 'Water', 'Yogurt'])
```

```python
def generate_database(items, filename):
    '''
    Creates database as pandas DataFrame and writes it to csv file
    Parameters:
        items (list): List of items that all customers are buying from
        filename (str): name of csv file you want to write to
    Returns:
        df (DataFrame): database of 20 transactions
    '''
    dataset = []
    for i in range(0, 20):
        j = random.randint(1, 10)
        transaction = random.sample(items, j)
        dataset.append(transaction)
    
    te = TransactionEncoder()
    te_ary = te.fit(dataset).transform(dataset)
    df = pd.DataFrame(te_ary, columns=te.columns_)
    
    # dict = {}
    # for item in items:
    #     values = [random.choice([True, False]) for i in range(20)]
    #     dict.update({item: values})
    # df = pd.DataFrame(dict)
    df.to_csv(filename)
    return(df)
```

```python
generate_database(items, 'Database1.csv')
generate_database(items, 'Database2.csv')
generate_database(items, 'Database3.csv')
generate_database(items, 'Database4.csv')
generate_database(items, 'Database5.csv')
```

In this database, the items are the columns, and the rows are the transactions. A True value indicates the items is part of the transaction, and a False value indicates the item is not part of the transaction.
